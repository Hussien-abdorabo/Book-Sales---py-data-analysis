```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz as sw
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
df = df[df['Publishing Year'] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset='Book Name',inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'],bins = 20)
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distributions of Publishing year')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of books')
plt.title('number of book in each genre')
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending = False).head()
```

```python
df.columns
```

```python
sns.boxplot(x="genre" ,y="Book_ratings_count", data =df)
plt.xlabel("Genre")
plt.ylabel("Books_Rating_Cont")
plt.title("Box Plot of Number of Book raing Count for each Genre")
plt.show()
```

```python
plt.scatter(df['sale price'], df['units sold'])
plt.xlabel(" Sale Price")
plt.ylabel(" units Sold")
plt.title("Scatter plot for sale price vs units sold")
```

```python
langauge_counts = df["language_code"].value_counts().sort_values(ascending= False)
```

```python
plt.pie(langauge_counts , labels = langauge_counts.index, startangle=90, autopct="%1.1f%%")
plt.title("pie")
plt.show()
```
